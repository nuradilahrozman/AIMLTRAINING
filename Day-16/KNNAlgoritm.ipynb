# KNN Algoritm in Supervised Machine Learning
# KNeighborsClassifier of sklearn library for this one

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Data collection
X = np.array([
    [1, 3500], 
    [2, 3200],
    [3, 3000],
    [4, 2500],
    [5, 2300],
    [6, 2000],
    [7, 2200],
    [8, 3700],
])
y =['Unifit','Unifit','Unifit','Unifit','Fit','Fit','Fit','Unifit']
```

```python
#create the model
model=KNeighborsClassifier(n_neighbors=3)
```

```python
#Train the Model
model.fit(X,y)
```

```python
#predict
new_person = [[4.5,2600]]
pred = model.predict(new_person)
print("Predicted Result:", pred[0])
```

```python
#predict from taking the values from users
exercise_hours=float(input('Enter Number of Exercise Hours'))
cal_intake=float(input('Enter Number of calories consumed'))
pred_fu=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories intake {cal_intake} your are likely {pred_fu[0]}')
```

## Download student_grade.cvs from github
## Predict a student's **grade level ( A,B OR C)**
## Based on ** hours studied** and **assignment completed**
## using the K-Nearest Neighbors algoritm 

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('student_grade.csv')
df
```

```python
le=LabelEncoder()
df['Grade']=le.fit_transform(df['Grade'])
X=df[['Hours_Studied','Assignments_Completed']]
y=df['Grade']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
hour_studied=float(input('Enter Hours of study'))
ass_completed=float(input('Enter Number of assignment completed'))
our_data=pd.DataFrame([[hour_studied,ass_completed]],columns=['Hours_Studied','Assignments_Completed'])
pred_grade=model.predict(our_data)
print(f'Hour of study : {hour_studied}, Number of assignment completed is : {ass_completed}. The grade is likely {pred_grade[0]}')
```
