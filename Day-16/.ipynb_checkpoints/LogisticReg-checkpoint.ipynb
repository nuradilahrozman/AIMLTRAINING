# Logistic Regression = Pass/Fail

```python
#import all required lib
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
#collect the data
df=pd.read_csv('pass_fail.csv')
df.head()
```

```python
#split the data
x = df[['Hours']]
y = df['Result']
```

```python
#create the model (change here only we are using LogisticRegression)
model=LogisticRegression()
```

```python
#trained the model
model.fit(x,y)
```

```python
pred_35 = model.predict([[3.5]])
print('Predicted result for 3.5 hours:', pred_35[0])
```

```python
hours_study=int(input('Number of Hours Studied'))
pred_value=model.predict([[hours_study]])
print(f'After {hours_study} hours study you are likely to {pred_value[0]}')
```

```python
#import all required lib
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('customer_purchase.csv')
df.head()
```

```python
#data split and labeling
le=LabelEncoder()
df['Purchased'] = le.fit_transform(df['Purchased'])
x=df[['Age','Salary']]
y=df['Purchased']
```

```python
df
```

```python
# creating the model
model=LogisticRegression()
```

```python
#Trained the model
model.fit(x,y)
```

```python
new_cust_age =int(input ('Enter Age'))
new_cust_sal =int(input ('Enter Salary'))
new_cust = [[new_cust_age, new_cust_sal]]
prediction = model.predict(new_cust)
if (prediction):
    print(f'Customer with age {new_cust_age} and salary {new_cust_sal} likely Purchased')
else:
    print(f'Customer with age {new_cust_age} and salary {new_cust_sal} likely not to Purchase')
```

# Logistic Regression - spam or not spam
## Download spam_data.csv from Kaggle.com
## Load the Data
## Do the data cleaning and labelling (if required) 
## Use LogisticRegression Algoritm to train the Model
## Predict Result if the email is spam or not spam


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('spam_dataset.csv')
df.head()
```

```python
X = df['message_content']
y = df['is_spam']
```

```python
#Convert text to numbers
cv = CountVectorizer()
X_cv = cv.fit_transform(X)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_cv, y)
```

```python
Sample = ['You have been pre-approved for a credit card!','Are you coming to class today']
pred = model.predict(cv.transform(Sample))
if(pred[0]):
    print('Not Spam')
else:
    print('Spam!!!')
```
