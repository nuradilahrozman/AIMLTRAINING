# supervised Machine Learning
## Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_csv('sales_data.csv')
data.head()
```

```python
x= data [["Advertising_Spend"]]# x must use 2 bracket to get the data cause it expect the 2d array
y = data ["Sales_Revenue"]
```

```python
model =LinearRegression()
model.fit(x,y)
print('Training Done')
```

```python
new_spend = float(input('Enter New Spend to Predict Sales Revenue'))
predicted_revenue = model.predict([[new_spend]])
print(f"New spend {new_spend} Predicted Revenue{predicted_revenue}")
```

## Linear Regression  - House Price Prediction
## Download house_price.csv file from github
## **Goal** Train a Linear Regression model to predict house price from size

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
df=pd.read_csv('house_price.csv')
df.head(2)
```

```python
house_size=df[['Size_sqft']]
house_price=df['Price']
print(house_size)
print(house_price)
```

```python
model= LinearRegression()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
model.fit(house_size,house_price)
```

```python
pr_value=model.predict([[1700]])
print(pr_value)
```
