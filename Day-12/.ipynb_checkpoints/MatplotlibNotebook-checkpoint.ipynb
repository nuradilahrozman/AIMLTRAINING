# Basic of Matplotlib chart
    ##Pyplot  :Interface for creating and managing plot
    ##Figure : The complete window to create plot
    ##Axes : The area where data is plotted
    ##Axis : x line and y line to define scale and ticks
    ##Plot : The graph of data

```python
plt.plot(x,y,marker='o',linestyle='-')
```

```python
plt.title('First line Chart')
```

```python
plt.xlabel('X label')
```

```python
plt.ylabel('Y label')
```

```python
plt.grid(True)
```

# First Example to create a line chart from our data
    ##student.xlsx for fetching the data
    ## Generate Line Chart based on student Name and student Marks

```python
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/Marqas/Desktop/AIMLTRAINING/Day-11/students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='s',color='pink')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(marks,names,marker='o',color='blue')
plt.title('Student Progress Chart')
plt.xlabel('Student Marks')
plt.ylabel('Student Name')
plt.show()
```

```python
print('Dash line')
plt.plot(marks,names,marker='o',color='blue',linestyle='--')
plt.show()
```

```python
print('Dotted line')
plt.plot(marks,names,marker='o',color='blue',linestyle=':')
plt.show()
```

```python
print('Dotted and dash line')
plt.plot(marks,names,marker='o',color='blue',linestyle='-.')
plt.show()
```
