# Scatter Plots
## In scatter plots we representdata in points, where each point represent a data observation. We can check how to values correlate 

```python
product_sales={'Laptop' :750,'Smart Phone':650,'Smart Watch':400,'Tablet':500,'AC':400}
product=list(product_sales.keys())
sales=list(product_sales.values())
plt.scatter(product,sales)
plt.title('Basic Product and Sales Representation')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
players=['kaka', 'wahid','kamarul','Faiz','syauqi']
match1_score=[45,67,56,79,89]
match2_score=[52,50,66,80,92]
match3_score=[48,70,50,88,79]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match one', s=100)
plt.scatter(players,match2_score,color='blue',label='Match one', s=100 , marker='d')
plt.scatter(players,match3_score,color='green',label='Match one', s=150 , marker='p')
plt.title('Comparison of Players in Three Matches')
plt.xlabel('Player')
plt.ylabel('Score')
plt.legend(loc='upper left')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

# Write a comparison that shows average calories burned for each workout type from final_data dataset

```python
df=pd.read_csv('Final_data.csv')
ave_cal=df.groupby('Workout_Type')['Calories_Burned'].mean()
selected_index=ave_cal.index
selected_values=ave_cal.values
plt.scatter(selected_index,selected_values,color='red', s=100)
plt.title('Comparison of Workout and calories burned')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```
