```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=100,color='green',edgecolor='black',alpha=0.8)
plt.title('Netflix released Movies and Show by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(color='lightblue',alpha=0.5,linestyle='--',linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=100,color='green',edgecolor='black',alpha=0.8)
plt.title('Netflix released Movies and Show by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(color='lightblue',alpha=0.5,linestyle='--',linewidth=0.5)
plt.tight_layout()
plt.savefig('netflix_bar.pdf',dpi=300,bbox_inches='tight')
plt.show()
```

```python
df=df.dropna(subset=['release_year','type'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
show=df[df['type']=='TV Show']
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'],show['release_year']],
         bins=30,
         color=['blue','red'],
         edgecolor='black',
         alpha=0.8,
         label=['Movies','TV Shows'])
plt.title('Netflix by Year and Type')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid(color='lightblue',alpha=0.5,linestyle='--',linewidth=0.7)
plt.tight_layout()
plt.legend(title='Movies/Shows')
plt.savefig('netflix_showvsmovies.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

# use netflix_data.csv dataset
### print how many Movies and TV shows are available on Netflix
### Display Movies and TV Shows proporpotion in a pie plot
### save files as pdf

```python
df=pd.read_csv('netflix_data.csv')
```

```python
df=df.dropna(subset=['type'])
```

```python
type_counts=df['type'].value_counts()
print(type_counts)
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(type_counts,labels=type_counts.index, colors=['pink','red'],autopct='%1.2f%%',startangle=145,shadow=True,explode=(0.05,0.05),radius=1.0)
plt.title('Comparison of Movies and Shows Number in Netflix')
plt.tight_layout()
plt.legend(title='Movies and show')
plt.savefig('netflix_showvsmovies.pdf',dpi=300,bbox_inches='tight')
plt.show()
```
