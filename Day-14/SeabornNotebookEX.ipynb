# Seaborn Library of Matplotlib
## Seaborn is a data visualization library in Matplotlib that provide high level interface for creating interactive and more colorful plots with less code

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palette=sns.color_palette()
print(palette)
```

```python
sns.palplot(palette)
```

```python
palette_green=sns.color_palette('Greens',20)
sns.palplot(palette_green)
```

```python
data_dict={
    'building':['burz-khalifa','medrika','klcc'],
    'category':['medium','low','highest'],
    'building_height':[150,120,165]
    }
bar_colors=['red','green','orange']
data=pd.DataFrame(data_dict)
sns.barplot(x='building',y='building_height',data=data,hue='category',palette=bar_colors)
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
data_iris
#data_flights=sns.load_dataset('flights')
#data_flights
```

```python
sns.barplot(x='species',y='sepal_width',data=data_iris,hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue',alpha=0.7)
```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue',alpha=0.7)
```

```python
sns.stripplot(x='species',y='sepal_width',data=data_iris,hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue',alpha=0.7)
```

```python
data_iris=sns.load_dataset('iris')
sns.pairplot(data=data_iris,hue='species')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Dark grid Theme Example Using Iris Data')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('White grid Theme Example Using Iris Data')
plt.show()
```

```python
sns.set_style('dark')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Dark Theme Example Using Iris Data')
plt.show()
```

```python
sns.set_style('white')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('White Theme Example Using Iris Data')
plt.show()
```

```python
#set_style can use darkgrid,whitegrid,dark,white or ticks
sns.set_style('ticks')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Ticks Theme Example Using Iris Data')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal Width in cm')
plt.xticks([4,5,6,7,8],fontsize=10,color='green')
plt.yticks([1,2,3,4,5,6,7],fontsize=10,color='purple')
sns.despine()
plt.show()
```
