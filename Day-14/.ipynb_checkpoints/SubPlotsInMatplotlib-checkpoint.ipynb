# Matplotlib.pyplot.Subplots()
## subplots(): provides a way to plot multiple plots on singles figure
## Given number of rows and column. It returns a tuple (fig,ax) giving a single figure fig with an array of axes ax

```python
import matplotlib.pyplot as plt
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
import numpy as np
```

```python
sales_data={'Electronics':45000,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
actual_sale=list(sales_data.values())
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(category,actual_sale)
ax4.hist(np.random.randn(1000))
```

# Subplots using indexing

```python
our_lines=np.linspace(0,10,100)
```

```python
product_prices={'Hair Oil':20,'Face Wash':30,'Lip Balm':10,'Face Cream':15}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(1000)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['blue','red','green','orange'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='green')
ax[1,1].hist(our_dots)
plt.tight_layout(rect=[0,0,1,0.95])# to adjust the numbering/xlabel distance
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['blue','red','green','orange'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='green')
ax[1,1].hist(our_dots)
#------------------------------------------
#First row 0 & first column 0 means [0,0]
ax[0,0].set_title('Line Plot Example')
ax[0,0].set_xlabel('X-Axis')
ax[0,0].set_ylabel('Y-Axis')
ax[0,0].grid(True,linestyle='--',alpha=0.7,color='Lightblue')
#-----------------------------------------
#First row 0 & second column 1 means [0,1]
ax[0,1].set_title('Bar Plot Example')
ax[0,1].set_xlabel('X-Axis')
ax[0,1].set_ylabel('Y-Axis')
ax[0,1].grid(True,linestyle='--',alpha=0.7,color='Lightgreen',axis='y')
#-----------------------------------------
#Second row 1 and First column 0 means [1,0]
ax[1,0].set_title('Our Scatter Plot Example')
ax[1,0].set_xlabel('Product Name')
ax[1,0].set_ylabel('Product Sale')
ax[1,0].grid(True,linestyle='--',alpha=0.7,color='Lightblue')
#-----------------------------------------
#second row 1 and second column 1 means [1,1]
ax[1,1].set_title('Histrogram Example')
ax[1,1].set_xlabel('Value Range')
ax[1,1].set_ylabel('Frequency')
ax[1,1].grid(True,linestyle='--',alpha=0.7,color='orange',axis='y')

fig.suptitle('Matplot Lib SubPlots example using 2 rows and 2 cols')
plt.tight_layout(rect=[0,0,1,0.95])# to adjust the numbering/xlabel distance
plt.savefig('subplotex.jpeg',dpi=300,bbox_inches='tight')
```
