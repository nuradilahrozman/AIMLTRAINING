```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/Marqas/Desktop/AIMLTRAINING/Day-13/netflix_data.csv')
df.head(3)
```

```python
sns.set_theme(style='darkgrid',palette='dark')
sns.countplot(data=df,x='type')
plt.title('Number of Movies Vs Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df.head(20),x='type',hue='country',palette='Set2')
plt.title('Netflix Content Type Country Only Top 10')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df,x='type',hue='country',palette='Set2')
plt.title('Netflix Content Type Country Only Top 10')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country',bbox_to_anchor=(1.02,1),loc='upper left',ncol=5,frameon=True)
plt.show()
```
