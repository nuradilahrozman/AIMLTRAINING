```python
import pandas as pd
```

```python
data={'Students':['Amit','Anam','Chi','Sam','Ravi','Javi'],
'Marks':[85,None,65,75,None,55]}
```

```python
df=pd.DataFrame(data)
print('Original Data')
df
```

```python
print('Missing Data')
print('Number of Missing values are:' ,df.isnull().sum())
```

```python
average_marks=df['Marks'].mean()
print(average_marks)
```

```python
df.fillna(average_marks, inplace=True)
```

```python
print('After filna() call')
df
```

# dropna() example

```python
Students_data={'Name':['Amit','Anam','Chi','Sam','Ravi','Javi'],
'Marks':[85,None,65,75,None,55]}
df=pd.DataFrame(Students_data)
print('Original Data')
df
```

```python
df=df.dropna(subset=['Marks'])
```

```python
print('Datset after dropna()')
df
```

```python
emp_data={
    'Name':['Sam','Ali','Chang','Xi','Viz','Suhaila'],
    'Age':[35,None,25,20,24,None],
    'Salary':[5000,4000,5000,4000,None,None],
}
df=pd.DataFrame(emp_data)
print('Original Data')
df
```

```python
print(df.isnull().sum())
```

```python
df_drop=df.dropna()
df_drop
```

```python
df_drop_salary=df.dropna(subset=['Salary'])
df_drop_salary
```

```python
#fillna example
df.fillna(df['Age'].mean(), inplace=True)
```

```python
df
```

```python
df.fillna({'Age':df['Age'].mean(),'Salary':df['Salary'].mean()}, inplace=True)
```
